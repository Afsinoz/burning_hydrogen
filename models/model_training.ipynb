```python
base_for_cv = XGBOnLinReg(device='gpu')
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_range = {
    'alpha': np.arange(0,5,0.1),
    'n_estimators': np.arange(10,500,1),
    'max_depth': np.arange(2,11,1),
    'learning_rate': np.arange(0.01, 0.1, 0.01),
    'subsample': np.arange(0.1, 1.0, 0.1)
}
```

```python
cv = RandomizedSearchCV(base_for_cv, param_distributions=param_range,
                        n_iter=10, scoring='neg_mean_squared_error',
                        n_jobs=-1, verbose=4, cv=3)
```

```python
cv.fit(X_train, y_train)
```

```python
cv.best_params_
```

```python
the_model = XGBOnLinReg(device='gpu', **cv.best_params_)
```

```python
the_model.fit(X_train, y_train)
```

```python
with open('/content/drive/MyDrive/Erdos/training/cv.pkl', 'ab') as f:
    pickle.dump(cv, f)
```

```python
with open('/content/drive/MyDrive/Erdos/training/best_model.pkl', 'ab') as f:
  pickle.dump(the_model, f)
```
